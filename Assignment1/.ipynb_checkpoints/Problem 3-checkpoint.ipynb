```python
# Import necessary libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

from matplotlib import pyplot as plt
import seaborn as sns

# Evaluation metric for regression
from sklearn.metrics import r2_score

%matplotlib inline
sns.set()
```

```python
# Utility functions

def compute_poly_features(X, n):
    '''Given a feature vector, it compute polynomial features and return matrix of dimension m x (n+1)'''
    m = X.shape[0]
    X = X.reshape((m, 1))
    x = np.empty((m, 0))
    
    for i in range(n+1):
        x = np.append(x, X**i, axis=1)
    
    return x

def get_weights(d, y):
    '''Compute coefficients of the nth degree polynomial for regression'''
    return np.matmul( np.matmul(np.linalg.inv(np.matmul(d.T, d)) , d.T), y)

def predict(X, w):
    '''Compute output given the test data'''
    return np.matmul(X,w)

def evaluate(y_pred, y_test):
    '''Evaluate the model using Mean Square Error and R2 Score'''
    mse = np.mean((y_test-y_pred)**2)
    r2score = r2_score(y_test, y_pred)
    return [mse, r2score] 

def plot_cost_function(X_train, y_train, X_test, y_test, min_dim=1, max_dim=10):
    '''Plot loss or cost function corresponding to each degree of polynomial'''
    J = []
    
    for dim in range(min_dim, max_dim+1):
        x = compute_poly_features(X_train, dim)
        w = get_weights(x, y_train)
        
        x_test = compute_poly_features(X_test, dim)
        y = predict(x_test, w)
        
        error = np.mean((y-y_test)**2)
        J.append(error)
        
    fig = plt.figure(figsize=(10,6))
    xtick = np.arange(min_dim, max_dim+1)
    plt.plot(xtick, J)
    plt.xticks(xtick)
    plt.xlabel('Degree of polynomial')
    plt.ylabel('Cost function')
    plt.title('J(w) vs Degree of polynomial')
    plt.show()
```

```python
# Reading csv files to pandas dataframe
df1 = pd.read_csv('Problem_3/Wage_original.csv', header=None)
df2 = pd.read_csv('Problem_3/Wage_dataset.csv', header=None)
```

```python
# Convert dataframe to numpy array
data = df2.as_matrix()

# 75-25 train-test split
data_train, data_test = train_test_split(data, test_size=0.25, random_state=42, shuffle=True)

# year, age, education, wage
X_train = data_train[:,[0,1,4]]
y_train = data_train[:, 9]

X_test = data_test[:,[0,1,4]]
y_test = data_test[:, 9]

m, n = X_train.shape

# As a sanity check, printing out sizes of training and test inputs
print ('Number of training samples \t: %d')%(m)
print 'Number of test samples \t\t: %d'%(X_test.shape[0])
print 'Dimension of data \t\t: %d'%(n)
```

## Visualizing input data

```python
fig = plt.figure(figsize=(18,9))
ax = fig.add_subplot(221)
ax.scatter(X_train[:,0], y_train, c=X_train[:,0], cmap='viridis')
ax.set_xlabel('year')
ax.set_ylabel('log(wage)')
ax.set_title('log(wage) vs year')
ax = fig.add_subplot(222)
ax.scatter(X_train[:,1], y_train, c=X_train[:,1], cmap='viridis')
ax.set_xlabel('age')
ax.set_ylabel('log(wage)')
ax.set_title('log(wage) vs age')
ax = fig.add_subplot(223)
ax.scatter(X_train[:,2], y_train, c=X_train[:,2], cmap='viridis')
ax.set_xlabel('education')
ax.set_ylabel('log(wage)')
ax.set_title('log(wage) vs education')
plt.show()
```

## Polynomial regression :      $ \hat{y} = \sum_{i=0}^{k} \omega_{i}x^{i} $

* Finding degree of polynomial at which loss is minimum.
* Getting coefficients of the polynomial.
* Evaluating the model using Mean Square Error and R$^{2}$ Score.

## Case 1 : Age vs Wage

```python
# Find optimal degree of polynomial for regression
plot_cost_function(X_train[:,1], y_train, X_test[:,1], y_test, min_dim=1, max_dim=10)

# Zooming in to get better view of the minima
plot_cost_function(X_train[:,1], y_train, X_test[:,1], y_test, min_dim=3, max_dim=9)
```

```python
degree = 7

# Computing polynomial features
X_train_age_poly = compute_poly_features(X_train[:,1], degree)
X_test_age_poly = compute_poly_features(X_test[:,1], degree)

# Get weights
w = get_weights(X_train_age_poly, y_train)

# Predict wage
y_pred = predict(X_test_age_poly, w)
```

```python
# Evaluate model
[mse, r2score] = evaluate(y_pred, y_test)
print "Mean Square error : %0.3f"%(mse)
print "R2 Score : %0.3f"%(r2score)
```

```python
# Plotting regression model
x = np.linspace(15,85,750)
x_poly = compute_poly_features(x, degree)
y = predict(x_poly, w)
fig = plt.figure(figsize=(10,6))
plt.scatter(X_test[:,1], y_test)
plt.xlabel('age')
plt.ylabel('log(wage)')
plt.plot(x, y, c='k')
plt.show()
```

## Case 2 : Education vs Wage

```python
# Find optimal degree of polynomial for regression
plot_cost_function(X_train[:,2], y_train, X_test[:,2], y_test, min_dim=1, max_dim=10)
plot_cost_function(X_train[:,2], y_train, X_test[:,2], y_test, min_dim=1, max_dim=4)
```

```python
degree = 3

# Computing polynomial features
x_train = X_train[:,2]
x_test = X_test[:,2]

# Get weights
x_train_poly = compute_poly_features(x_train, degree)
x_test_poly = compute_poly_features(x_test, degree)

# Get weights
w = get_weights(x_train_poly, y_train)

# Predict wage
y_pred = predict(x_test_poly, w)

# Evaluate model
[mse, r2score] = evaluate(y_pred, y_test)
print "Mean Square error : %0.3f"%(mse)
print "R2 Score : %0.3f"%(r2score)
```

```python
# Plotting regression model
x = np.linspace(min(x_train)-1, max(x_train)+1, x_test.shape[0])
x_poly = compute_poly_features(x, degree)
y = predict(x_poly, w)
fig = plt.figure(figsize=(10,6))
plt.scatter(x_test, y_test)
plt.xlabel('education')
plt.ylabel('log(wage)')
plt.plot(x, y, c='k')
plt.show()
```
