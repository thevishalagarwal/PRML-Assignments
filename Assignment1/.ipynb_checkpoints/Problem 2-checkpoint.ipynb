# Variation in performance of Bayesian Decision by different covariance matrix

The following problem is intended to illustrate alterations in performance and shape of distributions
brought about by variations in covariance matrices.

* Training data : `P2_train.csv` consisting of 310 instances, 2 attributes and 1 class label.
* Test data : `P2_test.csv` consisting of 310 instances, 2 attributes and 1 class label.

_Problem_: Learn a binary classifier for the given data taking class conditional densities as normal density. Estimate the misclassification rates of both classes, plot the discriminant function and iso-probability contours.

```python
# Import necessary libraries
import numpy as np
import pandas as pd
from scipy.stats import multivariate_normal

from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns

# Self-implemented APIs
from BayesClassifier import BayesClassifier
from discriminant_analysis import PCA
from discriminant_analysis import LDA

%matplotlib inline
sns.set()
```

```python
# Some utility functions

def plot_by_class_2d(X, y):
    i = np.where(y==0)[0]
    x1 = X[i,:]
    i = np.where(y==1)[0]
    x2 = X[i,:]

    fig = plt.figure(figsize=(10,6))
    plt.scatter(x1[:,0], x1[:,1], marker='^', alpha=0.6, label='Class 0')
    plt.scatter(x2[:,0], x2[:,1], marker='8', color='r', alpha=0.6, label='Class 1')
    plt.legend()
    
def plot_by_class_1d(X, y):
    i = np.where(y==0)[0]
    x1 = X[i,:]
    i = np.where(y==1)[0]
    x2 = X[i,:]

    fig = plt.figure(figsize=(10,6))
    plt.scatter(x1[:,0], np.zeros((x1.shape[0],)), marker='^', alpha=0.6, label='Class 0')
    plt.scatter(x2[:,0], np.zeros((x2.shape[0],)), marker='8', color='r', alpha=0.6, label='Class 1')
    plt.legend()
    
def initialize_val():
    '''Initialize value for plotting surface and contour'''
    x = np.linspace(-6,6,500)
    y = np.linspace(-6,6,500)

    x, y = np.meshgrid(x,y)

    z = np.empty(x.shape + (2,))
    z[:, :, 0] = x
    z[:, :, 1] = y
    
    return [x, y, z]
```

```python
# Reading csv files to pandas dataframe
df_train = pd.read_csv('P2_data/P2_train.csv', header=None)
df_test = pd.read_csv('P2_data/P2_test.csv', header=None)
```

```python
# Converting dataframe into matrices
X_train = df_train.values[:,:-1]
y_train = df_train.values[:,-1]
X_test = df_test.values[:,:-1]
y_test = df_test.values[:,-1]

m, n = X_train.shape

# As a sanity check, printing out sizes of training and test inputs

print ('Number of training samples \t: %d')%(m)
print 'Number of test samples \t\t: %d'%(X_test.shape[0])
print 'Dimension of data \t\t: %d'%(n)
```

## Visualizing training input

```python
plot_by_class_2d(X_train, y_train)
plt.show()
```

`BayesClassifier.py` contains the `BayesClassifier` class which contains the following methods :
- `fit()` which estimates parameters like mean, covariance and apriori probability using **Maximum Likelihood Estimation**
- `predict()` which predicts the output of given test data using the estimated parameters
- `evaluate()` which calculates the accuracy and confusion matrix for the model

_Assumption : The data from each class can be modelled using multivariate normal distribution._

```python
model = BayesClassifier()
# Estimating parameters
model.fit(X_train, y_train)
```

```python
# Getting parameters of our model after estimation
mean = model.mean
covariance = model.covariance
apriori = model.apriori
```

## Case 1 : Unequal covariance, $\Sigma_{5} \neq \Sigma_{6}$ 
Consider the general case of covariance matrix where off-diagonal elements are non-zero and the covariance matrix is of the form $\left(\begin{array}{cc} 
a & b\\
c & d
\end{array}\right)$

```python
# Initializing variables for surface plotting
x, y, z = initialize_val()

# Class likelihood
px_c0 = multivariate_normal.pdf(z, mean=model.mean[0,:], cov=model.covariance[0,:,:])
px_c1 = multivariate_normal.pdf(z, mean=model.mean[1,:], cov=model.covariance[1,:,:])

# Posterior Probability
pc0_x = px_c0 * apriori[0]
pc1_x = px_c1 * apriori[1]

# Discriminant function
g0 = -np.log(pc0_x)
g1 = -np.log(pc1_x)
g = g0-g1

# Plotting Class likelihood for each class
fig = plt.figure(figsize=(10,6), edgecolor='k')
ax = fig.gca(projection='3d')
ax.plot_surface(x, y, px_c0, rstride=3, cstride=3, linewidth=1, antialiased=True, cmap='viridis')
ax.plot_surface(x, y, px_c1, rstride=3, cstride=3, linewidth=1, antialiased=True, cmap='viridis')
plt.title('Multivariate Normal Distribution for 2 classes')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('P(X|C)')
plt.show()
```

```python
# Predict output using test set
y_pred_1 = model.predict(X_test)

# Evaluating the predicted output
acc_1, confusion_matrix_1 = model.evaluate(y_pred_1, y_test)
```

```python
# Display different evaluation metrics
print 'Overall Accuracy: %.2f%%\n'%(acc_1*100)
print 'Confusion Matrix : '
print confusion_matrix_1
```

## Case 2 : Equal covariance, $\Sigma_{5} = \Sigma_{6} = \Sigma = \left(\begin{array}{cc} 
a & b\\
c & d
\end{array}\right)$

In this case, consider a common covariance matrix for all classes. The common covariance matrix can be written as a sum of covariance matrix of each class weighted by the apriori probability of respective classes.

$\Sigma = \sum_{i} \hat{P}(C_i)\Sigma_{i}$

```python
# Computing single covariance class
for i in range(model.n_class):
    covariance[i,:,:] *= apriori[i]
covariance = np.sum(covariance, axis=0)

# Setting covariance attribute of model
model.covariance = covariance

# Predicting output on test data using single covariance matrix
y_pred_2 = model.predict(X_test)

# Evaluating the predicted output
acc_2, confusion_matrix_2 = model.evaluate(y_pred_2, y_test)
```

```python
# Initializing variables for surface plotting
x, y, z = initialize_val()

# Class likelihood
px_c0 = multivariate_normal.pdf(z, mean=model.mean[0,:], cov=model.covariance[:,:])
px_c1 = multivariate_normal.pdf(z, mean=model.mean[1,:], cov=model.covariance[:,:])

# Posterior Probability
pc0_x = px_c0 * apriori[0]
pc1_x = px_c1 * apriori[1]

# Discriminant function
g0 = -np.log(pc0_x)
g1 = -np.log(pc1_x)
g = g0-g1

# Plotting Class likelihood for each class
fig = plt.figure(figsize=(10,6), edgecolor='k')
ax = fig.gca(projection='3d')
ax.plot_surface(x, y, px_c0, rstride=3, cstride=3, linewidth=1, antialiased=True, cmap='viridis')
ax.plot_surface(x, y, px_c1, rstride=3, cstride=3, linewidth=1, antialiased=True, cmap='viridis')
plt.title('Multivariate Normal Distribution for 2 classes')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('P(X|C)')
plt.show()
```

## Case 3 : Diagonal covariance, $\Sigma_{5} = \Sigma_{6} = \Sigma = \left(\begin{array}{cc} 
a & 0\\
0 & b
\end{array}\right)$

In this case, we assume a common diagonal covariance matrix for all classes. The diagonal covariance matrix is derived by setting all off-diagonal elements as zero.

```python
# Computing diagonal covariance matrix
cova = model.covariance
cov = np.zeros((n,n))
cov[np.arange(n), np.arange(n)] = cova[np.arange(n), np.arange(n)]

# Setting covariance attribute of model
model.covariance = cov

# Predicting output on test data
y_pred_3 = model.predict(X_test)

# Evaluating the predicted output
acc_3, confusion_matrix_3 = model.evaluate(y_pred_3, y_test)
```

```python
# Initializing variables for surface plotting
x, y, z = initialize_val()

# Class likelihood
px_c0 = multivariate_normal.pdf(z, mean=model.mean[0,:], cov=model.covariance[:,:])
px_c1 = multivariate_normal.pdf(z, mean=model.mean[1,:], cov=model.covariance[:,:])

# Posterior Probability
pc0_x = px_c0 * apriori[0]
pc1_x = px_c1 * apriori[1]

# Discriminant function
g0 = -np.log(pc0_x)
g1 = -np.log(pc1_x)
g = g0-g1

# Plotting Class likelihood for each class
fig = plt.figure(figsize=(10,6), edgecolor='k')
ax = fig.gca(projection='3d')
ax.plot_surface(x, y, px_c0, rstride=3, cstride=3, linewidth=1, antialiased=True, cmap='viridis')
ax.plot_surface(x, y, px_c1, rstride=3, cstride=3, linewidth=1, antialiased=True, cmap='viridis')
plt.title('Multivariate Normal Distribution for 2 classes')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('P(X|C)')
plt.show()
```

## Case 4 : Equal diagonal covariance, $\Sigma_{5} = \Sigma_{6} = \Sigma = \left(\begin{array}{cc} 
a & 0\\
0 & a
\end{array}\right)$

We assume equal covariance in each feature and equal covariance matrix for each class. In this case, the covariance matrix becomes identity matrix, I, times the covariance and the Mahalanobis distance reduces to Euclidean distance.  
To get a scalar variance, dimensionality reduction is performed. The dimensionality techniques used are PCA and LDA and among these two LDA is found to perform well in this dataset. As shown below, LDA separates the class very well into two regions and hence to get the variance, variance after LDA is used.

```python
# Performing PCA on the dataset
pca = PCA()
pca.fit(X_train, n_components=1)
z1 = pca.transform(X_train)
plot_by_class_1d(z1, y_train)
plt.title('PCA')
plt.show()
```

```python
# Performing LDA on the dataset
lda = LDA()
lda.fit(X_train, y_train, n_components=1)
z2 = lda.transform(X_train)
plot_by_class_1d(z2, y_train)
plt.title('LDA')
plt.show()
```

```python
# Computing covariance
cov = np.cov(z2, rowvar=False)

# Setting covariance attribute of model
model.covariance = cov*np.eye(n)

# Predicting output on test data using single covariance matrix
y_pred_4 = model.predict(X_test)

# Evaluating the predicted output
acc_4, confusion_matrix_4 = model.evaluate(y_pred_4, y_test)
```

```python
# Initializing variables for surface plotting
x, y, z = initialize_val()

# Class likelihood
px_c0 = multivariate_normal.pdf(z, mean=model.mean[0,:], cov=model.covariance[:,:])
px_c1 = multivariate_normal.pdf(z, mean=model.mean[1,:], cov=model.covariance[:,:])

# Posterior Probability
pc0_x = px_c0 * apriori[0]
pc1_x = px_c1 * apriori[1]

# Discriminant function
g0 = -np.log(pc0_x)
g1 = -np.log(pc1_x)
g = g0-g1

# Plotting Class likelihood for each class
fig = plt.figure(figsize=(10,6), edgecolor='k')
ax = fig.gca(projection='3d')
ax.plot_surface(x, y, px_c0, rstride=3, cstride=3, linewidth=1, antialiased=True, cmap='viridis')
ax.plot_surface(x, y, px_c1, rstride=3, cstride=3, linewidth=1, antialiased=True, cmap='viridis')
plt.title('Multivariate Normal Distribution for 2 classes')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('P(X|C)')
plt.show()
```
